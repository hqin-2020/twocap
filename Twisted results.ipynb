# Section 1: One Capital Shock

# State Evolution 
$$
\begin{align*}
d \log K_t &= \left[D_t - \frac{\phi}{2}\left(D_t\right) ^ 2 + (.01)\left(\alpha_k + \beta Z_t \right) - {\frac {(.01) ^ 2 |\sigma_k | ^2}2} \right] dt + (.01)\sigma_k \cdot dW ^ H_t \\
    dZ_t &= \left(\alpha_z - \kappa Z_t \right)dt + \sigma_z \cdot dW ^ H_t \\
    dW_t &= H_t dt + d{W}_t^H
\end{align*}
$$

# Planner's Problem

The paper seek a function $V(z, k; \ell) = \log k + \nu(z; \ell)$, which delivers the robust planner's HJB equation as 

\begin{align*}
0 = & \max_{d} \min_{h} \ \ - \delta \nu(z; \ell) + \delta \log ({\mathcal A} - d) + \left[d - {\frac {\phi} 2} d^2 +  (.01) \left({\widehat \alpha}_k + {\widehat \beta} z - \frac{(.01)|\sigma_k^2|}{2}\right)\right] + (.01) \left[\sigma_k \cdot h\right]\cr
& \quad\quad + \frac{\partial \nu(z; \ell)}{ \partial z}\left( {\widehat \alpha}_z -{\widehat \kappa} z + \sigma_z \cdot h \right)  + \frac{1}{2}\frac{ \partial^2 \nu(z; \ell)}{\partial z^2} |\sigma_z|^2  \\
& \quad\quad  +  \frac{\ell}{2}\left[ |h|^2 - \xi(z) \right].
\end{align*}

# Planner's Optimal Choice

## Minimizing Choice of $h$

\begin{align*}
H^*(z; \ell) & \doteq \begin{bmatrix}h^*_k \\ h^*_z \end{bmatrix} = -\frac{1}{\ell}\left[ (.01) \sigma_k + \frac{\partial  \nu(z ; \ell)}{\partial z}\sigma_z  \right]. 
\end{align*}

## Maximizing Choice of $d$, Investment-Capital Ratio

\begin{align*}
 1 - \phi d^* =   {\frac \delta  {{\mathcal A} - d^*}}
\end{align*}

## Consumption-Capital Ratio

Let's otbain the optimal consumptaion-capital ratio $c := C/K$ as 

$$
c^*=\mathcal{A}-d^*
$$

# Section 2: Two Capital Shock

# State Evolution

Define $R_t=\exp(L_t)/(1+\exp(L_t))$, then state follow


\begin{align*}
d\log K_t &= \left[\varphi_1 (1-R_t) + \varphi_2 R_t - \frac{(.01)^2\left|\sigma_1(1-R_t) + \sigma_2 R_t\right|^2}{2}\right]dt + %\\
%& \hspace{1cm} +
(.01)\left[\sigma_1(1-R_t) + \sigma_2 R_t\right]\cdot dW_t \\
dL_t &= \left[\varphi_2 - \varphi_1 -  \frac{(.01)^2}{2}\left(|\sigma_2|^2 - |\sigma_1|^2 \right)\right] dt + (.01)\left[\sigma_2 -\sigma_1\right]\cdot dW_t \\
dZ_t &= - \widehat\kappa \left(Z_t - \bar z\right)dt + \sigma_z \cdot dW_t
\end{align*}

where 

$$
\varphi_i\left(D_t^{(i)}, Z_t\right) \doteq D_t^{(i)}-\frac{\phi_i}{2}\left(D_t^{(i)}\right)^2+(.01)\left(\widehat{\alpha}_z+\hat{\beta} Z_t\right)
$$
# Planner's Problem

The paper seek a value function  $V(X) = \log K + \nu(L, Z)$ that solves the HJB equation

\begin{align}
0 &= \max_{d^{(1)}, d^{(2)}}\min_{h}   - \delta\nu(l, z) + \ \delta\log\left((1-r)\left(\mathcal A_1 - d^{(1)}\right) + r\left(\mathcal A_2 - d^{(2)}\right) \right)  \\
& \quad\quad + \left[\varphi_1(1-r) + \varphi_2r - \frac{(.01)^2\left[\sigma_1(1-r) + \sigma_2 r\right]^2}{2} + (.01)[(1-r)\sigma_1 + r\sigma_2]\cdot h\right]   \\
& \quad\quad + \nu_l(l, z)\left[ \varphi_2 - \varphi_1 - \frac{(.01)^2}{2}\left(|\sigma_2|^2  - |\sigma_1|^2 \right) + (.01)[\sigma_2 - \sigma_1]\cdot h\right]  \\
& \quad\quad + \nu_z(l, z)\left[ -\widehat{\kappa}(z-\bar z) + \sigma_z \cdot h\right] + \frac{1}{2}\text{tr}\left(V_{xx}\sigma\sigma'\right) \\
& \quad\quad + \frac{\ell}{2}\left[|h|^2 - \xi(z)\right] 
\end{align}

where
\begin{align*}
\text{tr}\left(V_{xx}\sigma\sigma'\right) &= (.01)^2|\sigma_2-\sigma_1|^2 \nu_{ll}(l, z) +  2 (.01)\left([\sigma_2 - \sigma_1]\cdot \sigma_z\right)\nu_{lz}(l, z) +  |\sigma_z|^2 \nu_{zz}(l, z).
\end{align*}

# Planner's Optimal Choice

## Minimizing Choice of $h$


\begin{align*}
h(l, z, \ell^*) & =- \frac{1}{\ell^*} \sigma'(r) \begin{bmatrix}1 \\
\nu_l(l, z) \\
\nu_z(l, z)  \end{bmatrix}
\end{align*}


## Maximizing Choice of $d$, Investment-Capital Ratio
\begin{align}
\frac{\delta(1-r)}{(1-r)(\mathcal A_1 - d^{(1)}(l, z))+r(\mathcal A_2 - d^{(2)}(l, z))} &= \left(1-\phi_1d^{(1)}(l, z)\right)\left[1 - r - \nu_l(l, z)\right]  \\
\frac{\delta r}{(1-r)(\mathcal A_1 - d^{(1)}(l, z))+r(\mathcal A_2 - d^{(2)}(l, z))} &= \left(1-\phi_2d^{(2)}(l, z)\right)\left[r + \nu_l(l, z)\right]  \\
\end{align}


# Planner's Consumption and Investment Dynamics (* The Same)

## Consumption Dynamics

\begin{align}
\log C_t = \log K_t + \log([\mathcal{A}_1 - d^*_1(R_t, Z_t)](1-R_t) + [\mathcal{A}_2-d^*_2(R_t, Z_t)]R_t ).
\end{align}

## Total Investment Dynamics

Let's define $\log K_t \doteq \log\left(K^{(1)}_t+ K^{(2)}_t\right)$ then we have 
\begin{align*}
d\log K_t &= \left[\varphi_1 (1-R_t) + \varphi_2 R_t - \frac{(.01)^2\left|\sigma_1(1-R_t) + \sigma_2 R_t\right|^2}{2}\right]dt + %\\
%& \hspace{1cm} +
(.01)\left[\sigma_1(1-R_t) + \sigma_2 R_t\right]\cdot dW_t \\
\end{align*}

# Section 4: Two Capital Shock, Extended to $\rho\neq 1$

# State Evolution

Define $R_t=\exp(L_t)/(1+\exp(L_t))$, then state follow


\begin{align*}
d\log K_t &= \left[\varphi_1 (1-R_t) + \varphi_2 R_t - \frac{(.01)^2\left|\sigma_1(1-R_t) + \sigma_2 R_t\right|^2}{2}\right]dt + %\\
%& \hspace{1cm} +
(.01)\left[\sigma_1(1-R_t) + \sigma_2 R_t\right]\cdot dW_t \\
dL_t &= \left[\varphi_2 - \varphi_1 -  \frac{(.01)^2}{2}\left(|\sigma_2|^2 - |\sigma_1|^2 \right)\right] dt + (.01)\left[\sigma_2 -\sigma_1\right]\cdot dW_t \\
dZ_t &= - \widehat\kappa \left(Z_t - \bar z\right)dt + \sigma_z \cdot dW_t
\end{align*}

where 

$$
\varphi_i\left(D_t^{(i)}, Z_t\right) \doteq D_t^{(i)}-\frac{\phi_i}{2}\left(D_t^{(i)}\right)^2+(.01)\left(\widehat{\alpha}_z+\hat{\beta} Z_t\right)
$$
# Planner's Problem

The paper seek a value function  $V(X) = \log K + \nu(L, Z)$ that solves the HJB equation

\begin{align}
0 &= \max_{d^{(1)}, d^{(2)}}\min_{h}   \frac{\delta}{1-\rho} \left[ exp\left[ \left(1-\rho\right) \left(\log\left((1-r)\left(\mathcal A_1 - d^{(1)}\right) + r\left(\mathcal A_2 - d^{(2)}\right) \right) - v\left(z;\ell\right)\right) \right] - 1 \right] \\
& \quad\quad + \left[\varphi_1(1-r) + \varphi_2r - \frac{(.01)^2\left[\sigma_1(1-r) + \sigma_2 r\right]^2}{2} + (.01)[(1-r)\sigma_1 + r\sigma_2]\cdot h\right]   \\
& \quad\quad + \nu_l(l, z)\left[ \varphi_2 - \varphi_1 - \frac{(.01)^2}{2}\left(|\sigma_2|^2  - |\sigma_1|^2 \right) + (.01)[\sigma_2 - \sigma_1]\cdot h\right]  \\
& \quad\quad + \nu_z(l, z)\left[ -\widehat{\kappa}(z-\bar z) + \sigma_z \cdot h\right] + \frac{1}{2}\text{tr}\left(V_{xx}\sigma\sigma'\right) \\
& \quad\quad + \frac{\ell}{2}\left[|h|^2 - \xi(z)\right] 
\end{align}

where
\begin{align*}
\text{tr}\left(V_{xx}\sigma\sigma'\right) &= (.01)^2|\sigma_2-\sigma_1|^2 \nu_{ll}(l, z) +  2 (.01)\left([\sigma_2 - \sigma_1]\cdot \sigma_z\right)\nu_{lz}(l, z) +  |\sigma_z|^2 \nu_{zz}(l, z).
\end{align*}
